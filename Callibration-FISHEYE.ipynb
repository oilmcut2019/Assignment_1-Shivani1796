## How to use 3d printer. ?

- Repetier-Host is a free app that allows you to control your 3D printer. We have used "Repetier-Host Windows 2.1.6."

- First step is to connect 

- Then go to "Object Placement" & click on (+) icon to add your file (3d object)

- After adding image file, Go to "Slicer"

- |> "Slice with CuraEngine"

- Go to "Manual Control" and maintain the:

1) Bed Temperature till 75 degree &

2) Extruder temperature till 195 degree

- Lastly, go to "Print Preview"  and "print"
